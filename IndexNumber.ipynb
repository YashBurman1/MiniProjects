```python
import pandas as pd
import numpy as np
import math as m
```

```python
df = pd.read_csv("/content/Stats_Practical_DataSet.csv")
df
```

```python
def Index_table(df_,B_Year,Year):
  result = pd.DataFrame()
  lst_p1q0 = []
  lst_p0q0 = []
  lst_p1q1 = []
  lst_p0q1 = []
  lst_EM_N = []
  lst_EM_D = []
  lst_EM_N_Q = []
  lst_EM_D_Q = []
  for i in df_.index:
    lst_p1q0.append(float(df_["Price_"+ Year][i]*df_["Quantity_" + B_Year][i]))
    lst_p0q0.append(float(df_["Price_"+B_Year][i]*df_['Quantity_'+ B_Year][i]))
    lst_p1q1.append(float(df_["Price_"+Year][i]*df_["Quantity_"+Year][i]))
    lst_p0q1.append(float(df_["Price_"+B_Year][i]*df_["Quantity_"+Year][i]))
  for i in df_.index:
    lst_EM_N.append(float(lst_p1q0[i]+lst_p1q1[i]))
    lst_EM_D.append(float(lst_p0q0[i]+lst_p0q1[i]))
    lst_EM_N_Q.append(float(lst_p0q1[i]+lst_p1q1[i]))
    lst_EM_D_Q.append(float(lst_p0q0[i]+lst_p1q0[i]))

  result["Commodity"]= ["Rice","jowar","Bazra","Wheat","Barley","Gram","Total"]
  lst_p1q0.append(float(sum(lst_p1q0)))
  lst_p0q0.append(float(sum(lst_p0q0)))
  lst_p1q1.append(float(sum(lst_p1q1)))
  lst_p0q1.append(float(sum(lst_p0q1)))
  lst_EM_N.append(float(sum(lst_EM_N)))
  lst_EM_D.append(float(sum(lst_EM_D)))
  lst_EM_N_Q.append(float(sum(lst_EM_N_Q)))
  lst_EM_D_Q.append(float(sum(lst_EM_D_Q)))
  result["p1q0"] = lst_p1q0
  result["p0q0"] = lst_p0q0
  result["p1q1"] = lst_p1q1
  result["p0q1"] = lst_p0q1
  result["p1(q0+q1)"] = lst_EM_N
  result["p0(q0+q1)"] = lst_EM_D
  result["q1(p0+p1)"] = lst_EM_N_Q
  result["q0(p0+q1)"] = lst_EM_D_Q
  result.set_index("Commodity",inplace = True)
  return result,float(sum(lst_p1q0)),float(sum(lst_p0q0)),float(sum(lst_p1q1)),float(sum(lst_p0q1)),float(sum(lst_EM_N)),float(sum(lst_EM_D)),float(sum(lst_EM_N_Q)),float(sum(lst_EM_D_Q))
```

```python
yrs = ["1983","1984",'1985']
Indices = {}
Indices["INDEX"] = ["Lapeyres Price Index","Paasches Price Index","Fishers Price Index","Edgeworth Marshall Price Index","Lapeyres Qty. Index","Paasches Qty. Index","Fishers Qty. Index","Edgeworth Marshall Qty. Index"]
for i in yrs:
  table,P1Q0,P0Q0,P1Q1,P0Q1,EM_N_P,EM_D_P,EM_N_Q,EM_D_Q = Index_table(df,"1982",i)
  las_P = (P1Q0/P0Q0)*100
  Pas_P = (P1Q1/P0Q1)*100
  Fis_P = m.sqrt((las_P * Pas_P))
  EM_P = (EM_N_P/EM_D_P)*100
  las_Q = (P0Q1/P0Q0)*100 
  Pas_Q = (P1Q1/P1Q0)*100
  Fis_Q = m.sqrt((las_Q * Pas_Q))
  EM_Q = (EM_N_Q/EM_D_Q)*100
  Indices[i] = [las_P,Pas_P,Fis_P,EM_P,las_Q,Pas_Q,Fis_Q,EM_Q]
  table = table.style.set_caption('Calculation of Indices for the Year' + i + " with 1982 as the Base year")
  #table.to_excel('Calculation of Indices for the Year' + i + " with 1982 as the Base year.xlsx")
  display(table)
fin_result = pd.DataFrame(Indices)
fin_result.set_index("INDEX",inplace = True)
fin_result = fin_result.style.set_caption('SUMMARY')
display(fin_result)
#fin_result.to_excel("SUMMARY.xlsx")
```
